```python
import pandas as pd


# Load the two CSV files
csv1 = pd.read_csv('context_augmentation.csv')  #['User Input', 'sentiment','Restaurant', 'Link', 'Output Reviews','Context Augmentation']
csv2 = pd.read_csv('consistency.csv')  #Consistency
csv3 = pd.read_csv('AI_tone_score.csv')  #AI-tone Score
csv4 = pd.read_csv('overall_meaningfulness.csv')  #overall meaningfulness&coherency(1-5)

csv2_selected = csv2[['Consistency']]
csv3_selected = csv3[['AI-tone Score']]
csv4_selected = csv4[[' overall meaningfulness&coherency(1-5)']]
df = pd.concat([csv1, csv2_selected,csv3_selected,csv4_selected], axis=1)
df = df[['User Input', 'sentiment','Restaurant', 'Link', 'Output Reviews','Consistency','Context Augmentation','AI-tone Score',' overall meaningfulness&coherency(1-5)']]

df.head()
```

```python
# avg score for each criterion
avg_consistency = df['Consistency'].mean()
avg_context_augmentation = df['Context Augmentation'].mean()
avg_ai_tone_score = df['AI-tone Score'].mean()
avg_meaningfulness_coherency = df[' overall meaningfulness&coherency(1-5)'].mean()

print(f"Average Consistency: {avg_consistency}")
print(f"Average Context Augmentation: {avg_context_augmentation}")
print(f"Average AI-tone Score: {avg_ai_tone_score}")
print(f"Average Overall Meaningfulness & Coherency: {avg_meaningfulness_coherency}")
```

```python
# overall success rate
columns_to_check = ['Consistency', 'Context Augmentation', 'AI-tone Score', ' overall meaningfulness&coherency(1-5)']
df['Failure Count'] = (df[columns_to_check] < 3).sum(axis=1)
num_rows_with_failures = (df['Failure Count'] > 0).sum()
print(f"Number of fauilre: {num_rows_with_failures}")
```

```python
df.to_csv('eval_result.csv', index=False)
```
